**data standardization to change units for our convention**

```python
df["highway-l/100km"] = 235/df["highway-mpg"]
```

**normalization is breaking the large numbers in small ranges(0-1)**

```python
df["length"] = df["length"]/df["length"].max()
```

```python
df["width"] = df["width"]/df["width"].max()
```

**binning is a process in we catogrize the large numbers in 3 ranges low medium high**

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_name =["low","medium","high"]
```

```python
df["horsepower_binned"] = pd.cut(df["horsepower"],bins, labels = group_name, include_lowest = True )
```

```python
df["horsepower_binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_name, df["horsepower_binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

**indicator variable or dummy variable we make that to provide easiness to our models.    We see the column "fuel-type" has two unique values: "gas" or "diesel". Regression doesn't understand words, only numbers. To use this attribute in regression analysis, we convert "fuel-type" to indicator variables.**

```python
dummy = pd.get_dummies(df["fuel-type"])
```

```python
dummy.head()
```

```python
dummy.rename(columns={"gas":"gas type_fuel","diesel":"diesel type_fuel"},inplace =True)
df.head(20)
```

```python
df = pd.concat([df,dummy],axis = 1)
```

```python
del df["fuel-type"]
```
